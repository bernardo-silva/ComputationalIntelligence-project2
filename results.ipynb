```python
from TSP_SO import SingleObjectiveTSP
from TSP import plot_route
import numpy as np
import matplotlib.pyplot as plt
```

```python
dist_central = np.loadtxt("CustDist_WHCentral.csv", skiprows=1, delimiter=",")[:,1:].tolist()
xy_central   = np.loadtxt("CustXY_WHCentral.csv", skiprows=1, delimiter=",")[:,1:].T

dist_corner  = np.loadtxt("CustDist_WHCorner.csv", skiprows=1, delimiter=",")[:,1:].tolist()
xy_corner    = np.loadtxt("CustXY_WHCorner.csv", skiprows=1, delimiter=",")[:,1:].T

orders       = np.loadtxt("CustOrd.csv", skiprows=1, delimiter=",")[:,1].tolist()
```

# 10 costumers 

```python
N=10
```

## 10 customers, WH central, orders file

```python
EA = SingleObjectiveTSP(ind_size = N,
                        distances= dist_central,
                        orders   = orders,
                        coords   = xy_central.T[1:N+1],
                        pop_size = 40,
                        use_heuristic= False,
                        elitist_cross= 20, 
                        elitist_size = 3,
                        CXPB  = 0.8,
                        MUTPB = 0.2,
                        INVPB = 0.8)

result10WHCentralOrdFile = EA.many_runs(n_runs=30)

print(f"Best fitness out of 30 runs: {result10WHCentralOrdFile['best_fitness']}")
print(f"Mean fitness out of 30 runs: {np.mean(result10WHCentralOrdFile['final_fitnesses'])}")
print(f"Std fitness out of 30 runs: {np.std(result10WHCentralOrdFile['final_fitnesses'])}")

plot_route(result10WHCentralOrdFile["best_solution"], xy_central.T[:N+1], orders[:N+1])
```

## 10 customers, WH central, orders 50

```python
EA = SingleObjectiveTSP(ind_size = N,
                        distances= dist_central,
                        orders   = [0] + [50]*N,
                        coords   = xy_central,
                        pop_size = 40,
                        use_heuristic = False,
                        elitist_cross = 20, 
                        elitist_size  = 3,
                        CXPB  = 0.8,
                        MUTPB = 0.2,
                        INVPB = 0.8)

result10WHCentralOrd50 = EA.many_runs(n_runs=30)

print(f"Best fitness out of 30 runs: {result10WHCentralOrd50['best_fitness']}")
print(f"Mean fitness out of 30 runs: {np.mean(result10WHCentralOrd50['final_fitnesses'])}")
print(f"Std fitness out of 30 runs: {np.std(result10WHCentralOrd50['final_fitnesses'])}")

plot_route(result10WHCentralOrd50["best_solution"], xy_central.T[:N+1], orders[:N+1])
```
