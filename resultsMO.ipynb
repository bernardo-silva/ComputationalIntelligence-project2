```python
from TSP_MO import MultipleObjectiveTSP
from TSP import plot_route, plot_route_with_labels
import numpy as np
import matplotlib.pyplot as plt
```

```python
dist_central = np.loadtxt("CustDist_WHCentral.csv", skiprows=1, delimiter=",")[:,1:].tolist()
xy_central   = np.loadtxt("CustXY_WHCentral.csv", skiprows=1, delimiter=",")[:,1:].T

orders       = np.loadtxt("CustOrd.csv", skiprows=1, delimiter=",")[:,1].tolist()
```

# 10 customers, WH central, orders file

```python
N=10
EA10 = MultipleObjectiveTSP(ind_size = N,
                        distances= dist_central,
                        orders   = orders,
                        coords   = xy_central.T[1:N+1],
                        pop_size = 100,
                        CXPB  = 0.8,
                        INVPB = 0.8)
```

```python
hv10, pop10 = EA10.run_algorithm()

dist10 = [x.fitness.values[0] for x in pop10]
cost10 = [x.fitness.values[1] for x in pop10]

print(f"Min distance: {min(dist10)}")
print(f"Min cost: {min(cost10)}")
print(f"Hypervolume: {hv10[-1]}")
```

```python
fig, ax = plt.subplots(figsize=(7,3.5),ncols=2)
ax[0].set(xlabel="Distance", ylabel="Cost")
ax[1].set(xlabel="Generation", ylabel="Hypervolume")
ax[0].scatter(dist10, cost10)
ax[1].plot(hv10)
ax[0].grid()
ax[1].grid()
fig.tight_layout()
```
