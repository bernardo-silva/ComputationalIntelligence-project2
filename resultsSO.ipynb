```python
from TSP_SO import SingleObjectiveTSP
from TSP import plot_route, plot_route_with_labels
import numpy as np
import matplotlib.pyplot as plt
```

```python
dist_central = np.loadtxt("data/CustDist_WHCentral.csv", skiprows=1, delimiter=",")[:,1:].tolist()
xy_central   = np.loadtxt("data/CustXY_WHCentral.csv", skiprows=1, delimiter=",")[:,1:].T

dist_corner  = np.loadtxt("data/CustDist_WHCorner.csv", skiprows=1, delimiter=",")[:,1:].tolist()
xy_corner    = np.loadtxt("data/CustXY_WHCorner.csv", skiprows=1, delimiter=",")[:,1:].T

orders       = np.loadtxt("data/CustOrd.csv", skiprows=1, delimiter=",")[:,1].tolist()
```

# 10 costumers 

```python
N=10
```

## 10 customers, WH central, orders file

```python
EA = SingleObjectiveTSP(ind_size = N,
                        distances= dist_central,
                        orders   = orders,
                        coords   = xy_central.T[1:N+1],
                        pop_size = 40,
                        use_heuristic= False,
                        elitist_cross= 20, 
                        elitist_size = 3,
                        CXPB  = 0.8,
                        INVPB = 0.8)

result10WHCentralOrdFile = EA.many_runs(n_runs=30)

print(f"Best fitness out of 30 runs: {result10WHCentralOrdFile['best_fitness']}")
print(f"Mean fitness out of 30 runs: {np.mean(result10WHCentralOrdFile['final_fitnesses'])}")
print(f"Std fitness out of 30 runs: {np.std(result10WHCentralOrdFile['final_fitnesses'])}")

plot_route(result10WHCentralOrdFile["best_solution"], xy_central.T[:N+1], orders[:N+1])
```
