```python
from TSP_SO import many_runs
from TSP import plot_route
import numpy as np
import matplotlib.pyplot as plt
```

```python
dist_central = np.loadtxt("CustDist_WHCentral.csv", skiprows=1, delimiter=",")[:,1:]
xy_central   = np.loadtxt("CustXY_WHCentral.csv", skiprows=1, delimiter=",")[:,1:].T

dist_corner  = np.loadtxt("CustDist_WHCorner.csv", skiprows=1, delimiter=",")[:,1:]
xy_corner    = np.loadtxt("CustXY_WHCorner.csv", skiprows=1, delimiter=",")[:,1:].T

orders       = np.loadtxt("CustOrd.csv", skiprows=1, delimiter=",")[:,1]
```

# 10 costumers

## 10 customers, WH central, orders file

```python
(best_solution, best_fitness, best_evolution_CentralOrdFile), \
(_, final_fitnesses) = many_runs(
                            ind_size=10,
                            distances=dist_central,
                            orders=orders,
                            pop_size=30,
                            elitist_cross=3, 
                            elitist_size = 30//5,
                            CXPB = 1,
                            MUTPB = 0.2,
                            INVPB=0.3,
                            n_runs=30)

print(f"Best fitness out of 30 runs: {best_fitness}")
print(f"Mean fitness out of 30 runs: {np.mean(final_fitnesses)}")
print(f"Std fitness out of 30 runs: {np.std(final_fitnesses)}")

plot_route(best_solution, xy_central.T[:11], orders[:11])
```
